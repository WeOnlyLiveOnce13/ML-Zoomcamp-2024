### Split the Dataset

* Split your data into training, validation, and test sets with a distribution of 60%/20%/20%.
* Use Scikit-Learn's `train_test_split` function for this task and set the random seed to **42**.
* Ensure that the target variable `y` is not included in your DataFrame.

```python
from sklearn.model_selection import train_test_split

df_train_full, df_test = train_test_split(df, test_size=0.2, random_state=42)
df_train, df_val = train_test_split(df_train_full, test_size=0.25, random_state=42)

len(df_train),len(df_val),len(df_test)
```

Let us remove **y** from the dataframe

```python
## ---- Extract and create Y for the different splits: ------
y_train = df_train.y.values
y_val = df_val.y.values
y_test = df_test.y.values

### ------ Delete Y from the splits: -------
del df_train['y']
del df_val['y']
del df_test['y']
```

### Question 3

* Calculate the mutual information score between `y` and other categorical variables in the dataset. Use the training set only.* 
Round the scores to 2 decimals using round(score, 2)
.
Which of these variables has the biggest mutual information score?

```python
## ----- Define categotical variables: ------------
categorical = ['job', 'marital', 'education', 'housing', 'contact', 'month', 'poutcome']
```

```python
from sklearn.metrics import mutual_info_score

### ----- Define the function to calculate the M.I on the training set ONLY
def calculate_mi(series):
    return mutual_info_score(series, y_train)

#### ---- Calculate MI between 'y' and the categorical variables of the training set ONLY 
df_mi = df_train[categorical].apply(calculate_mi)
df_mi = df_mi.sort_values(ascending=False).to_frame(name='MI')
```

```python
print('Below are the variable with highest M.I score:')
display(df_mi.head(n=10))
```

### Question 4

* Now let's train a logistic regression.* 
Remember that we have several categorical variables in the dataset. Include them using one-hot encoding.* 
Fit the model on the training dataset* .
To make sure the results are reproducible across different versions of Scikit-Learn, fit the model with these parameter    " s:
mode**l = LogisticRegression(solver='liblinear', C=1.0, max_iter=1000, random_state** "=* 42)
Calculate the accuracy on the validation dataset and round it to 2 decimal digits.

```python
#### One-hot encoding of the categorical features
from sklearn.feature_extraction import DictVectorizer
```

```python
## ----- Initialize the encoder: ------
dv = DictVectorizer(sparse=False)

## ---- Apply the transformation on the training set
train_dict = df_train[categorical + numerical].to_dict(orient='records')
X_train = dv.fit_transform(train_dict)

## ---- Apply the transformation on the validation set (for evaluation)
val_dict = df_val[categorical + numerical].to_dict(orient='records')
X_val = dv.transform(val_dict)
```

```python
#### Training on Logistic regression
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(solver='liblinear', C=1.0, max_iter=1000, random_state = 42)
model.fit(X_train, y_train)
```

```python
### Let us evaluate on the Validation set
y_pred = model.predict_proba(X_val)[:, 1]

### Set a threshold for our loan decision
loan_decision = (y_pred >= 0.5)

### ---- Accuracy of the model ----
accuracy = (y_val == loan_decision).mean()

print('Accuracy on the validation set is', round(accuracy, 2), '%.')
```

### Question 6
Smallest `C` that leads to the best accuracy on the validation set

```python
from sklearn.linear_model import LogisticRegression

for C in [0.01, 0.1, 1, 10, 100]:
    # Initialize the Logistic Regression model with the current value of C
    model = LogisticRegression(solver='liblinear', C=C, max_iter=1000, random_state=42)
    
    # Fit the model on the training data
    model.fit(X_train, y_train)

    ### ---- Get the prediction probabilities
    y_pred = model.predict_proba(X_val)[:, 1]

    ### Set a threshold for our loan decision
    loan_decision = (y_pred >= 0.5)

    ### ---- Calculate accuracy of the model ----
    accuracy = (y_val == loan_decision).mean()

    ### ---- Print different Accuracy ---
    print('Accuracy on the validation set for C = %.2f is %.2f%%.' % (C, accuracy * 100))
```

> Conclusion: Accuracy increases as we increase C from 0.01 up to 1. However, it seems to have reached a peak at 1, then decreases from 1 to 10, and increases again to 100.
`C=1` is therefore the smallest C.
